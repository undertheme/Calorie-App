```python
nutrients1 = nutrients1.loc[:, ['name', 'calories', 'total_fat', 'cholesterol', 'irom', 'sugars', 'carbohydrate']]
```

```python
def clean_and_convert(value):
    if isinstance(value, str):
        if 'mg' in value:
            value = ''.join(c for c in value if c.isdigit() or c == '.')
            value = float(value) / 1000  # Convert mg to g
        elif 'g' in value:
            value = ''.join(c for c in value if c.isdigit() or c == '.')
            value = float(value)
        else:
            value = float(value)
    return value

# Apply the function to each relevant column
nutrients1['fat'] = nutrients1['fat'].apply(lambda x: clean_and_convert(x))
nutrients1['iron'] = nutrients1['iron'].apply(lambda x: clean_and_convert(x))
nutrients1['sugars'] = nutrients1['sugars'].apply(lambda x: clean_and_convert(x))
nutrients1['cholesterol'] = nutrients1['cholesterol'].apply(lambda x: clean_and_convert(x))
nutrients1['carbohydrate'] = nutrients1['carbohydrate'].apply(lambda x: clean_and_convert(x))
```

```python
nutrients2['fat'] = (
    nutrients2['Data.Fat.Monosaturated Fat'] +
    nutrients2['Data.Fat.Polysaturated Fat'] +
    nutrients2['Data.Fat.Saturated Fat'] +
    nutrients2['Data.Fat.Total Lipid']
)
```

```python
## In above dataset we are also missing calories. Calories include carbohydrates, Protein, Fat, Alcohol
## I'm using Atwater system to calculate the calories based on 100g serving.

## formular for calories: calories = (
#     4 * carbs_g + 4 * protein_g + 9 * fat_g
# )
nutrients2['calories'] = (
    4 * nutrients2['Data.Carbohydrate'] +
    4 * nutrients2['Data.Protein'] +
    9 * nutrients2['fat']
)
```
